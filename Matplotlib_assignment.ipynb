#### Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.

#### Matplotlib :
Matplotlib is a widely-used data visualization library in Python that allows users to create a wide variety of static, interactive, and animated plots. It was originally created by John D. Hunter in 2003 and is now maintained by the Matplotlib development team

#### Why it Used :
Matplotlib is a powerful tool for data visualization that serves several essential purposes:

1. Data Exploration :
----------------------
Matplotlib enables users to quickly visualize their data, making it easier to understand patterns, trends, and outliers in the data. This helps in exploring data before diving into in-depth analysis.

2. Data Communication : 
-------------------------------
Plots and visualizations are an effective way to communicate insights and results to others. Matplotlib allows users to create publication-quality figures suitable for presentations, reports, and research papers.

3. Interactivity : 
---------------------
Matplotlib supports interactivity, allowing users to zoom, pan, and interact with plots, enhancing the data exploration experience.

4. Customization : 
---------------------------
The library provides a high degree of customization, allowing users to control almost every aspect of the plots, such as colors, line styles, markers, labels, titles, and annotations.

5. Integration :
-----------------------
Matplotlib is widely used in conjunction with other Python libraries for data analysis and scientific computing, such as NumPy, Pandas, and SciPy.


#### 1. Line Plot :

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 40, 6, 80, 10]

plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot')
plt.show()
```

#### 2. Scatter Plot :

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [200, 4, 600, 8, 1000]

plt.scatter(x, y,c='r')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot')
plt.show()
```

#### 3. Bar Plot :

```python
import matplotlib.pyplot as plt

categories = ['A', 'B', 'C']
values = [10, 150, 120]

plt.bar(categories, values)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot')
plt.show()
```

#### 4. Histogram :

```python
import matplotlib.pyplot as plt

data = [2, 5, 3, 8, 4,2,2,3,3,5]

plt.hist(data)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()
```

#### 5. Pie Chart:

```python
import matplotlib.pyplot as plt

categories = ['A', 'B', 'C', 'D']
values = [25, 35, 20, 40]

plt.pie(values, labels=categories, autopct='%1.1f%%')
plt.title('Pie Chart')
plt.show()
```

### subplot() :
Matplotlib's subplot() function allows you to create multiple plots arranged in a grid-like structure within a single figure. This can be particularly useful when you want to compare multiple plots or visualize different aspects of your data side by side.

```python
import numpy as np
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([0, 100, 200, 300, 400, 500])

plt.subplot(2, 2, 1)
plt.plot(x,y)



x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([50, 20, 40, 20, 60, 70])

plt.subplot(2, 2, 2)
plt.plot(x,y)


x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([10, 20, 30, 40, 50, 60])


plt.subplot(2, 2, 3)
plt.plot(x,y)

x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([200, 350, 250, 550, 450, 150])

plt.subplot(2, 2, 4)
plt.plot(x,y)

plt.show()
```

#### Box Plot : 
A Box Plot is also known as Whisker plot is created to display the summary of the set of data values having properties like minimum, first quartile, median, third quartile and maximum. In the box plot, a box is created from the first quartile to the third quartile, a vertical line is also there which goes through the box at the median. Here x-axis denotes the data to be plotted while the y-axis shows the frequency distribution.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
plt.boxplot(box1)
```

```python
plt.boxplot(box2)
```
